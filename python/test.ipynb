```python
import matplotlib.pyplot as plt
import math
def cal_similarity_of_vect(a, b):
    dot_product = a[0] * b[0] + a[1] * b[1]
    magnitude_a = (a[0] ** 2 + a[1] ** 2) ** 0.5
    magnitude_b = (b[0] ** 2 + b[1] ** 2) ** 0.5
    if magnitude_a == 0:
        magnitude_a = 1e-20
    if magnitude_b == 0:
        magnitude_b = 1e-20
    cosine_of_angle = dot_product / (magnitude_a * magnitude_b)
    arccos = math.acos(cosine_of_angle)
    return (math.pi - arccos) / math.pi

# Example usage
vector1 = [1, 1]
vector2 = [-1, -1]
similarity = cal_similarity_of_vect(vector1, vector2)
print(similarity)
plt.figure(figsize=(6, 6))
# 画出向量
plt.quiver(0, 0, vector1[0], vector1[1], angles='xy', scale_units='xy', scale=1, color='r')
# plt.legend()
plt.quiver(0, 0, vector2[0], vector2[1], angles='xy', scale_units='xy', scale=1, color='b')
#自动调整刻度
plt.axis('equal')
plt.xlim(-2, 2)
plt.ylim(-2, 2)
plt.show()
```
